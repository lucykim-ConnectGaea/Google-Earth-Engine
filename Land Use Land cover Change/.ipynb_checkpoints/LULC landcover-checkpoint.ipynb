## Load Image and filter 2010-2020


```python
Map = geemap.Map()
collection = (
    ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
    .filterDate('2015-01-01', '2022-01-01')
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 5))
)
image = collection.median()

vis = {
    'min': 0.0,
    'max': 3000,
    'bands': ['B4', 'B3', 'B2'],
}

Map.setCenter(37.083375, -1.038757, 12)
Map.addLayer(image, vis, 'Sentinel-2')
Map
```

## Number of images included in the collection after applying the filter

```python
count = collection.size().getInfo()
```

```python
print("the number of images is:", count)
```

## Select the best image each year

```python
start_year = 2018
end_year = 2022
years = ee.List.sequence(start_year, end_year)
year_list = years.getInfo()
print(year_list)
```

```python
roi = ee.Geometry.Point(37.083375, -1.038757)  # Nairobi , Kenya


def best_image(year):
    start_date = ee.Date.fromYMD(year, 1, 1)
    end_date = start_date.advance(1, 'year')
    
    image = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED') \
        .filterBounds(roi) \
        .filterDate(start_date, end_date) \
        .sort("CLOUD_COVER") \
        .first()
    
    return image
```

```python
images = years.map(best_image)
```
